# Netflix Exploratory Data Analysis

## Objective
Analyze the Netflix Movies and TV Shows dataset to uncover trends in content type, ratings, genres, countries, and release years using Python.

## Tools Used
- Python
- Pandas
- NumPy
- Matplotlib

## Dataset
Netflix Movies and TV Shows Dataset (Kaggle)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("netflix_titles_nov_2019.csv")
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
#movies vs tv shows
df["type"].value_counts()
```

```python
df["type"].value_counts().plot(kind="bar")

plt.title("Movies vs TV Shows")

plt.xlabel("Type")

plt.ylabel("Count")

plt.show()
```

```python
#top countries
df["country"].value_counts().head(10)
```

```python
df["country"].value_counts().head(10).plot(kind="bar")

plt.title("Top 10 Countries")

plt.xlabel("Country")

plt.ylabel("Number of Shows")

plt.xticks(rotation=45)

plt.show()
```

```python
#top 10 ratings
df["rating"].value_counts()
```

```python
df["rating"].value_counts().plot(kind="bar")

plt.title("Ratings Distribution")

plt.show()
```

```python
#content added per year
df["release_year"].value_counts().sort_index().plot()

plt.title("Content Released by Year")

plt.xlabel("Year")

plt.ylabel("Count")

plt.show()
```

```python
#top 10 directors
df["director"].value_counts().head(10)
```

```python
df["director"].value_counts().head(10).plot(kind="bar")

plt.title("Top Directors")

plt.show()
```

```python
#top 10 genres
df["listed_in"].value_counts().head(10)
```

```python
df["listed_in"].value_counts().head(10).plot(kind="bar")

plt.title("Top Genres")

plt.show()
```

# Conclusion

### Key Insights

- Movies are more common than TV Shows.
- The United States has the largest number of Netflix titles.
- TV-MA is one of the most common content ratings.
- Netflix expanded its content library significantly over recent years.
- Drama and International Movies are among the most popular genres.
